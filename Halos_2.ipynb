# Top 100 Halos

```python
top_100 = pd.read_csv('Halos_2/merged/top_100_merged.csv')
top_100
```

```python
top_100_old = pd.read_csv('Halos/merged/top_100_merged.csv')
top_100_old
```

```python
kt = 'fof_halo_specific_angmom_mag_%s'
ms100 = 5
ms10 = 10
ms1 = 20
plt.plot(top_100_old[kt%'orig'],top_100_old[kt%'comp'],'.b',markersize=ms100)
plt.plot(top_100_old[kt%'orig'][0:10],top_100_old[kt%'comp'][0:10],'.k',markersize=ms10)
plt.plot(top_100_old[kt%'orig'][0],top_100_old[kt%'comp'][0],'.r',markersize=ms1)
plt.xlabel(kt%'orig')
plt.ylabel(kt%'comp')
plt.title('old')
plt.show()
plt.plot(top_100[kt%'orig'],top_100[kt%'comp'],'.b',markersize=ms100)
plt.plot(top_100[kt%'orig'][0:10],top_100[kt%'comp'][0:10],'.k',markersize=ms10)
plt.plot(top_100[kt%'orig'][0],top_100[kt%'comp'][0],'.r',markersize=ms1)
plt.xlabel(kt%'orig')
plt.ylabel(kt%'comp')
plt.title('new')
plt.show()
```

```python
for k in ['fof_halo_count_orig','fof_halo_specific_angmom_mag_orig']:
    print('NEW')
    print(top_100[k].describe(percentiles=[]))
    print('')
    print('OLD')
    print(top_100_old[k].describe(percentiles=[]))
    print('')
```

# All Halos Above 10,000 Particles

```python
above_10000 = pd.read_csv('Halos_2/merged/above_10000_merged.csv')
above_10000.sort_values(by=['mag_diff_orig'],ascending=False,inplace=True)
above_10000
```

```python
above_10000_old = pd.read_csv('Halos/merged/above_10000_merged.csv')
above_10000_old.sort_values(by=['mag_diff_orig'],ascending=False,inplace=True)
above_10000_old
```

```python
k = 'mag_diff_orig'
plt.plot(above_10000_old[k],'b',label='old')
plt.plot(above_10000[k],'r',label='new')
plt.yscale('log')
plt.ylabel(k)
plt.legend()
plt.show()
```

```python
kx = 'fof_halo_specific_angmom_mag_orig'
ky = 'mag_diff_orig'
plt.plot(above_10000_old[kx],above_10000_old[ky],'.b',label='old')
plt.plot(above_10000[kx],above_10000[ky],'.r',label='new')
plt.yscale('log')
plt.xscale('log')
plt.xlabel(kx)
plt.ylabel(ky)
plt.legend()
plt.show()
```
